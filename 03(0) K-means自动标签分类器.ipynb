```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from collections import Counter

#
file = open('data.txt', mode='r')
str = file.readline().split(", ")
raw_data_x = []
for i in range(len(str)):
    location = []
    if i == 0:
        location = [str[i], str[i+1].split()[0]]
    elif i == len(str)-1:
        break
    else:
        location = [str[i].split()[1], str[i+1].split()[0]]
    raw_data_x.append(location)
raw_data_x = np.array(raw_data_x,dtype='float32')    
#print(raw_data_x) 
raw_data_x_label = np.full([raw_data_x.shape[0]],-1)
num_data = raw_data_x.shape[0]
#print(num_data)
plt.scatter(raw_data_x[:,0],raw_data_x[:,1])


#k
#np.random.seed(666)
k=2
center_index = np.random.choice(num_data,k,replace=False)#
center = raw_data_x[center_index]
#print(center)
for cent in center:
    plt.scatter(cent[0],cent[1],color='r')
#plt.axis([0,10,0,8])
plt.show()
label = 0
for cent_index in center_index:#
    raw_data_x_label[cent_index] = label
    label+=1
#print(raw_data_x_label)    

def k_means(center):
    #
    n=1 #
    distances = []
    for x in raw_data_x:#
        d = [(np.sum(abs(x - cent) ** n)**(1/n)) for cent in center]
        distances.append(d)

    #print(distances)

    #
    for i in range(0,num_data):
        nearest = np.argsort(distances)
    #print(nearest)

    for i in range(0,k):#
        raw_data_x_label[nearest[:,i]==0] = i
    #print(raw_data_x_label)

    #
    new_center = []
    for j in range(0,k):
        d = [np.mean(raw_data_x[raw_data_x_label==j,i])for i in range(0,raw_data_x.shape[1])]
        new_center.append(d)
    #print(new_center)
    return new_center


sum=0
while sum!=num_data:
    old_raw_data_x_label = np.copy(raw_data_x_label)
    center = k_means(center)
    new_raw_data_x_label = np.copy(raw_data_x_label)
    sum = np.sum(old_raw_data_x_label==new_raw_data_x_label)
    print(sum)
    plt.figure()
    for cent in center:
        plt.scatter(cent[0],cent[1],color='r')
    for i in range(0,k):
        plt.scatter(raw_data_x[raw_data_x_label==i,0],raw_data_x[raw_data_x_label==i,1])
    #plt.axis([0,10,0,8])
    plt.show()
        
```
